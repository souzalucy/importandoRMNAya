#IMPORTANDO BIBLIOTECA E BASE DE DADOS

```python
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

#separando em treino e teste

```python
x_treinamento, X_teste, y_treinamento, y_teste = train_test_split(iris.data, iris.target, test_size=0.3, random_state=0)
```

#construindo e aplicando o modelo

```python
modelo = MLPClassifier(
    verbose=True,
    hidden_layer_sizes=(5,4,5),
    activation= 'relu',
    batch_size = 20,
    learning_rate='adaptive',
    momentum=0.9,
    early_stopping=False,
    max_iter=1000,
    random_state=10
)
```

```python
modelo.fit(x_treinamento, y_treinamento)
```

```python
plt.plot(modelo.loss_curve_)
plt.xlabel("interações")
plt.ylabel("valor de loss")
plt.title('loss')
```

```python
previsoes = modelo.predict(X_teste)
previsoes
```

```python
print(accuracy_score(y_teste,previsoes))
```
